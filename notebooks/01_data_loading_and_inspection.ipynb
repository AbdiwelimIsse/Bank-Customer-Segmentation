# Phase 1 & 2: Data Loading and Inspection
This notebook covers the initial loading of the dataset and verification of its integrity before any preprocessing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options for better visibility
pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")
```

## 1. Data Loading
The dataset is semi-colon (`;`) delimited. We use the full 45,211 row dataset.

```python
# Load the full dataset
file_path = '../data/raw/bank-full.csv'
df = pd.read_csv(file_path)

# Verify shape
print(f"Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")
df.head()
```

## 2. Data Inspection
Checking for nulls, duplicates, and data types.

```python
# Check data types and nulls
df.info()
```

```python
# Check for missing values explicitly
missing_values = df.isnull().sum().sum()
print(f"Total Missing Values: {missing_values}")

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Total Duplicate Rows: {duplicates}")
```

## 3. Statistical Summary of Target Features
Reviewing the 5 key features: `age`, `balance`, `duration`, `campaign`, `previous`.

```python
target_features = ['age', 'balance', 'duration', 'campaign', 'previous']
df[target_features].describe()
```
